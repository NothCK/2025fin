```python
import talib
import yfinance as yf
```

```python
df = yf.download("GOOGL")
df
```

```python
import numpy as np
import pandas as pd
```

```python
Close = df['Close']['GOOGL'].to_numpy()

RSI = talib.RSI(Close,timeperiod=14)
```

```python
buy_list = []
sell_list = []

for i in range(len(df)):
    buy = 0
    sell = 0
    if RSI[i] > 70:
        sell = 1
    if RSI[i] < 30:
        buy = 1
    buy_list.append(buy)
    sell_list.append(sell)
```

```python
RSI_df = pd.DataFrame(RSI, columns=["RSI"], index=df.index)
buy_df = pd.DataFrame(buy_list, columns=["Buy_Signal_RSI"], index=df.index)
sell_df= pd.DataFrame(sell_list, columns=["Sell_Signal_RSI"], index=df.index)
df_merged = pd.concat([df,RSI_df,buy_df,sell_df], axis=1, join="inner" )
df_merged
```
